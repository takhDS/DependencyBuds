```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

edges = pd.read_csv(r"../data/edges_updated_reversed.csv", sep=' ') #load the edges
edges_array = np.array(edges.loc[:100, ['Source', 'Target']]) #convert it to np.array

G = nx.DiGraph() #empty DiGraph
G.add_edges_from(edges_array)
```

```python
# pos = nx.spring_layout(G, seed=8020)
# nx.draw(G, pos, with_labels=True)

for i in nx.simple_cycles(G):
    print(i)
# No cycles!
```
